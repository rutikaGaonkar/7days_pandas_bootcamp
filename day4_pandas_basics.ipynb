```python
import pandas as pd 
data = { 
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fay', 'George'],
    'Salary': [50000, 55000, 48000, 51000, 60000, 62000, 68000],
    'Experience': [2, 3, 1, 2, 4, 5, 3]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.sort_values('Salary')
```

```python
df.sort_values('Salary', ascending=False)
```

```python
df.sort_values(by=['Department', 'Salary'], ascending=[True, False])
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Employee'].count()
```

```python
df.groupby('Department').agg({
'Salary': ['mean', 'max', 'min'],
'Experience': 'mean'
})
```

```python
df.groupby('Department').agg(
    avg_salary = ('Salary', 'mean'),
    max_salary = ('Salary', 'max'),
    min_salary = ('Salary', 'min'),
    avg_experience = ('Experience', 'mean')
)
```

```python
df.sort_values('Experience', ascending=False)
```

```python
df.groupby('Department')['Experience'].mean()
df.groupby('Department').agg(
avg_experience = ('Experience', 'mean'),
)
```

```python
avg_salary = df.groupby('Department')['Salary'].mean()

highest_avg_dept = avg_salary.idxmax()
highest_avg_salary = avg_salary.max()

print(f"The department with the highest average salary is {highest_avg_dept} with an average of {highest_avg_salary}.")
```

```python
df.groupby('Department').agg(
    total_salary = ('Salary', 'sum'),
    avg_experience = ('Experience', 'mean')
)
```
