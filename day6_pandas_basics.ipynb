```python
import pandas as pd
```

```python
data = {
    'Date': ['2025-07-01', '2025-07-02', '2025-07-03', '2025-07-04', '2025-07-05'],
    'Sales': [100, 150, 200, 170, 130]
}

df=pd.DataFrame(data)
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.dtypes) #to confirm the 'Date' column is now datetime
```

```python
df.set_index('Date', inplace=True)
print(df)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
print(df)
```

```python
filtered_df = df[df.index > '2025-07-03']
print(filtered_df)
```

```python
weekly_sales = df['Sales'].resample('W').sum()
print(weekly_sales)
```

```python
import matplotlib.pyplot as plt
```

```python
df['Sales'].plot(title='Daily Sales Over Time', figsize = (8, 4))
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
weekly_sales.plot(title='Weekly Sales Resampled', figsize = (8, 4), marker = 'o')
plt.xlabel=('Week')
plt.ylabel=('Sales')
plt.grid(True)
plt.show()
```
